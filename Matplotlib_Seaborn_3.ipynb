```python
#Write a Python program to count the number of strings having length more than 2 and are palindrome in a list using loop.
#Input:- ['ab', 'abc', 'aba', 'xyz', '1991']
#Output:- 2

inp_lst = ['ab', 'abc', 'aba', 'xyz', '1991']
pal_count = 0
for i in inp_lst:
    if i == i[::-1]:
        pal_count += 1
print(pal_count)

```
