# Matplotlib and seaborn

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arr1 = np.arange(0,5,0.4)
arr1
```

```python
arr2 = np.log(arr1)
arr2
```

```python
plt.plot(arr1, arr2, 'r')
plt.xlabel('arr1')
plt.ylabel('arr2')
plt.title('arr1 vs arr2')
```

```python
plt.subplot(1,2,1)
plt.plot(arr1, arr2, 'r')
plt.subplot(1,2,2)
plt.plot(arr2, arr1, 'g')
plt.subplot(1,2,1)
plt.plot(arr1, arr1*2, 'b')
plt.subplot(1,2,2)
plt.plot(arr1, arr1*3, 'r')
```

# Object oriented methods of plotting the graph

```python
cnv = plt.figure()
ax = cnv.add_axes([0.1, 0.1, 0.8, 0.8])
# Plotting arr1 vs arr2
ax.plot(arr1, arr2, 'r', marker = 'o')
ax.set_xlabel('Arr1')
ax.set_ylabel('Arr2')
ax.set_title('Arr1 vs Arr2')
```

```python
cnv = plt.figure()
ax1 = cnv.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = cnv.add_axes([0.5, 0.2, 0.3, 0.3])
# Plotting arr1 vs arr2
ax1.plot(arr1, arr2, 'r', marker = 'o')
ax1.set_xlabel('Arr1')
ax1.set_ylabel('Arr2')
ax1.set_title('Arr1 vs Arr2')

ax2.plot(arr2, arr1, 'g', marker = 's')
ax2.set_xlabel('Arr1')
ax2.set_ylabel('Arr2')
ax2.set_title('Arr2 vs Arr1')
```

```python
cnv, ax = plt.subplots(nrows = 3, ncols = 3)
ax[0, 0].plot(arr1, arr2, 'r')
ax[0,1].plot(arr2, arr1, 'g')
plt.tight_layout()
```

```python
cnv = plt.figure(figsize=(10,5))

```

```python
cnv, ax = plt.subplots(figsize=(10,5), dpi = 50)
ax.plot(arr1, arr2, 'r')
```

```python
cnv.savefig('log_graph.png')
```

```python
# Setting legend
cnv = plt.figure()
ax = cnv.add_axes([0.1, 0.1, 0.8, 0.8])
# Plotting arr1 vs arr2
ax.plot(arr1, arr2, 'r', marker = 'o', label = 'Arr1 vs Arr2')
ax.set_xlabel('Arr1')
ax.set_ylabel('Arr2')
ax.set_title('Arr1 vs Arr2')
ax.legend()
```

```python
cnv = plt.figure()
ax1 = cnv.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = cnv.add_axes([0.5, 0.2, 0.3, 0.3])
# Plotting arr1 vs arr2
ax1.plot(arr1, arr2, 'r--', marker = 'o', label = 'Arr1 vs Arr2', alpha = 1, lw = 0.1)
ax1.set_xlabel('Arr1')
ax1.set_ylabel('Arr2')
ax1.set_title('Arr1 vs Arr2')
ax1.legend()

ax2.plot(arr2, arr1, 'g.-', marker = 's', label = 'Arr2 vs Arr1')
ax2.set_xlabel('Arr1')
ax2.set_ylabel('Arr2')
ax2.set_title('Arr2 vs Arr1')

ax2.legend()
```

```python
cnv = plt.figure()
ax = cnv.add_axes([0.1, 0.1, 0.8, 0.8])
# Plotting arr1 vs arr2
ax.plot(arr1, arr2, 'r', marker = 'o', label = 'Arr1 vs Arr2', ls = ':')
ax.set_xlabel('Arr1')
ax.set_ylabel('Arr2')
ax.set_title('Arr1 vs Arr2')
```

```python
cnv = plt.figure()
ax = cnv.add_axes([0.1, 0.1, 0.8, 0.8])
# Plotting arr1 vs arr2
ax.plot(arr1, arr2, 'g', marker = 'o',markersize = 8, markerfacecolor = 'Yellow', markeredgecolor = 'green', markeredgewidth = 2)
ax.set_xlabel('Arr1')
ax.set_ylabel('Arr2')
ax.set_title('Arr1 vs Arr2')

# Marker = +, o, *, s, 1, 2, 3, 4....
```

# Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pok_data = pd.read_csv('Datasets/Pokemon.csv')
```

```python
pok_data.head()
```

```python
pok_data.describe()
```

```python
# sns.distplot() wiull give histogram and kde curve both for a univariate plot
grph = sns.distplot(pok_data['Health Point'], color = 'Green')
grph.set_title('Health Point vs Frequency')
grph.set_xlabel('Health Point')
grph.set_ylabel('Frequency')
```

```python
sns.distplot(pok_data['Health Point'], kde = False)
```

```python
# sns.distplot() wiull give histogram and kde curve both for a univariate plot
grph = sns.distplot(pok_data['Health Point'], color = 'red', hist = False)
grph.set_title('Health Point vs Frequency')
grph.set_xlabel('Health Point')
grph.set_ylabel('Frequency')
```

```python
# rug plot inside distplot
grph = sns.distplot(pok_data['Health Point'], rug = True)
grph.set_title('Health Point vs Frequency')
grph.set_xlabel('Health Point')
grph.set_ylabel('Frequency')
```

```python
sns.rugplot(pok_data['Health Point'], axis = 'x', height = 0.5)
```

```python
sns.distplot(pok_data['Health Point'], kde = False, rug = True, bins = 100)
```

```python
sns.distplot(pok_data['Health Point'], hist = False, rug = True, bins = 100)
```

```python
sns.distplot(pok_data['Health Point'], hist = False, rug = True, bins = 10)
```

```python
sns.set(color_codes = True)
sns.kdeplot(pok_data['Health Point'], shade = True, color = 'green')
sns.rugplot(pok_data['Health Point'], axis = 'x')
```

```python
# plotting bivariate data
sns.jointplot(x = 'Health Point', y = 'Total Point', data = pok_data)
```

```python
sns.jointplot(x = 'Health Point', y = 'Total Point', data = pok_data, xlim= {0,500}, ylim = {0,1000})
```
