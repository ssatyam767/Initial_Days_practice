# Iteration over an array

```python
arr1 = np.array([10,20,30,40,50])
for i in arr1:
    print(i)
```

```python
arr2 = np.arange(25, 55).reshape(5, 6)
arr2
```

```python
for i in arr2:
    print(i)
```

```python
for row in arr2:
    for ele in row:
        print(ele)
```

```python
arr2
```

```python
for ele in np.nditer(arr2[3], order = 'C'):
    print(ele)
    
# order = 'C' is called row major form
```

```python
for ele in np.nditer(arr2, order = 'F'):
    print(ele)

# order = 'F' is called column major form
```

# Splitting an array

```python
arr1 = np.arange(20, 32)
arr1
```

```python
np.split(arr1, 3)
```

```python
np.split(arr1, 4)
```

```python
# np.split(arr1, 5)
```

```python
np.split(arr1, [2, 6, 10])
```

```python
arr2 = np.arange(25, 61).reshape(6,6)
arr2
```

```python
ar1, ar2 = np.split(arr2, 2, axis = 0)   # If axis attriute is not provided then by default it will be splitting along row
print(ar1)
print()
print(ar2)
```

```python
ar1, ar2 = np.split(arr2, 2, axis = 1)   # If axis attriute is not provided then by default it will be splitting along row
print(ar1)
print()
print(ar2)
```

```python
ar1, ar2, ar3 = np.split(arr2, 3, axis = 0)   # If axis attriute is not provided then by default it will be splitting along row
print(ar1)
print()
print(ar2)
print()
print(ar3)
```

```python
arr2
```

```python
ar1, ar2, ar3, ar4 = np.split(arr2, [1, 3, 4], axis = 0)   # If axis attriute is not provided then by default it will be splitting along row
print(ar1)
print()
print(ar2)
print()
print(ar3)
print()
print(ar4)
```

```python
ar1, ar2, ar3, ar4 = np.split(arr2, [1, 3, 4], axis = 1)
print(ar1)
print()
print(ar2)
print()
print(ar3)
print()
print(ar4)
```

# Concatenation of an array

```python
arr1 = np.array([['UP', 'MP'], ['KA', 'TN']])
arr2 = np.array([['LKO', 'BHP'], ['BLR', 'CHN']])
print(arr1)
print(arr2)
```

```python
np.concatenate((arr1, arr2), axis = 0)
```

```python
np.concatenate((arr1, arr2), axis = 1)
```

# Image manipulation using numpy

```python
import numpy as np
from skimage import data
```

```python
astronaut = data.astronaut()
```

```python
astronaut
```

```python
astronaut.size
```

```python
astronaut.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(astronaut)
```

```python
cropped_image = astronaut[:, 50:, :]
plt.imshow(cropped_image)
```

```python
im1, im2 = np.split(astronaut, 2, axis = 0)
plt.imshow(im1)
```

```python
plt.imshow(im2)
```

```python
imv1, imv2 = np.split(astronaut, 2, axis = 1)
plt.imshow(imv1)
```

```python
plt.imshow(imv2)
```

# Joining the images

```python
joined_img = np.concatenate((im1, im2))
plt.imshow(joined_img)
```

```python
joined_img = np.concatenate((imv1, imv2),axis = 1)
plt.imshow(joined_img)
```
