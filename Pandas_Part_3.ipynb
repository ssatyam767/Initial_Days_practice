# Iterate over a dataframe

```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
                   'Headquarters' : ["Paris", "Bangalore", "Washington",
                                     "Tokyo", "Beijing"],
                   'Country_name' : ["France", "India", "USA", "Japan", "China"],
                 'Formation_year' : [1961, 1969, 1958, 2003, 1993]
             }
```

```python
df = pd.DataFrame(space_org)
```

```python
df
```

```python
for i in df:
    print(i)
```

```python
for i in df.itertuples():
    print(i)
```

```python
for i in df.iterrows():
    print(i[0])
    print(i[1]['Organizations_name'])
    print()
```

```python
for i in df.iteritems():
    print(i[0])
    print(i[1])
    print()
```

# Concatenation and merging dataframe

```python
students = pd.DataFrame({'id_number' : [204, 202, 264, 136, 238],
                               'Name' : ['James Smith',
                                        'Maria Rodriguez',
                                        'David Smith', 
                                        'Patricia Garcia',
                                        'Jessica Hernandez'],
                               'Age' : [26, 22 ,24, 21, 25],
                            'Gender' : ['M', 'F', 'M', 'F', 'F'],
                    
                        })
```

```python
students
```

```python
professors = pd.DataFrame({'Prof_name' : ['Thomas Smith',
                                          'Margaret Garcia',
                                          'James Johnson', 
                                          'Richard Miller',
                                          'Margaret MOORE'],
                      'id_number' : [204, 264, 108, 238, 136],
                     'Department' : ['ECE', 'EE' ,'CSE', 'ME', 'CE'],
                      
                         })
```

```python
professors
```

```python
pd.concat([students, professors], sort = True, ignore_index = True)
```

```python
pd.merge(students, professors)
```

```python
pd.merge(students, professors, on = 'id_number')
```

```python
pd.merge(students, professors, on = 'id_number', how = 'left')
```

```python
pd.merge(students, professors, on = 'id_number', how = 'right').replace(np.NaN, 'ABC')
```

# Sorting dataframe

```python
movie_rating = pd.read_csv('datasets/unsorted_imdb_rating.csv')
movie_rating
```

```python
movie_rating.index = [9,2,8,7,1,6,3,5,4,0]
```

```python
movie_rating
```

```python
movie_rating.sort_index(ascending = False)
```

```python
movie_rating.sort_values(by = 'duration')
```
