## Loading dataset

```python
# Importing library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setting figuresize and dpi
plt.rcParams['figure.figsize'] = [8,4]
plt.rcParams['figure.dpi'] = 100
```

```python
# Loading titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
cols = titanic.columns
cols
```

```python
titanic.describe()
```

```python
titanic.info()
```

# Heatmap

```python
# Setting up plot style
plt.style.use('ggplot')
```

```python
titanic.isnull()           # Gives a masked array where True represents values is null and False represents value is False
```

```python
titanic.isnull().sum()        # Given how many values are null in every column
```

```python
sns.heatmap(titanic.isnull(), cmap = 'viridis', cbar = True)''
```

```python
# Plotting the heatmap of all corelated dataset - i.e for all the datapoints which are numerical datapoints
# Creating corelation matrix 
cormat = titanic.corr()
cormat
```

```python
sns.heatmap(cormat, annot = True, fmt = '.2f', cmap = 'Reds', vmax = 0.8)
```

#### Note:- Heatmap is not recommended to find the corelation between categorical datapoints

# Univariate Analysis

```python
cnv, ax = plt.subplots(3,3, figsize = (20,20))
sns.countplot('survived', data = titanic, ax = ax[0][0])
sns.countplot('pclass', data = titanic, ax = ax[0][1])
sns.countplot('sex', data = titanic, ax = ax[0][2])
sns.countplot('parch', data = titanic, ax = ax[1][0])
sns.countplot('alone', data = titanic, ax = ax[1][1])
sns.countplot('sibsp', data = titanic, ax = ax[1][2])
sns.distplot(titanic['fare'], ax = ax[2][0])
sns.distplot(titanic['age'], ax = ax[2][1])
sns.countplot('embarked', data = titanic, ax = ax[2][2])
plt.tight_layout()
```
