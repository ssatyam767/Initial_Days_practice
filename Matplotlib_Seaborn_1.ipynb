```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr1 = np.arange(25, 35)
arr1
```

```python
arr2 = arr1**2
arr2
```

```python
plt.plot(arr1, arr2)
plt.xlabel('Arr1')
plt.ylabel('Arr2')
plt.title('Arr1 vs Arr2')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 0.8, 0.8])

ax.plot(arr1, arr2)
ax.set_xlabel('Arr1')
ax.set_ylabel('Arr2')
ax.set_title('Arr1 vs Arr2')
```

```python
apple = pd.read_csv('datasets/apple_stocks.csv')
high = np.array(apple['High'])
low = np.array(apple['Low'])
open1 = np.array(apple['Open'])
fig = plt.figure()

ax = fig.add_axes([0, 0, 0.8, 0.8])

ax.plot(apple['High'], apple['Low'])
ax.set_xlabel('High')
ax.set_ylabel('Low')
ax.set_title('High vs low')
plt.show()
```

```python
# Color, ls, lw
apple = pd.read_csv('datasets/apple_stocks.csv')
high = np.array(apple['High'])
low = np.array(apple['Low'])
open1 = np.array(apple['Open'])
close1 = np.array(apple['Close'])
volume = np.array(apple['Volume'])


fig, axes = plt.subplots(nrows = 2, ncols =2, figsize = (10,8))
axes[0][0].plot(high, low, color = 'Red', ls = ':', lw = 1)
axes[0][0].set_xlabel('High')
axes[0][0].set_xlabel('Low')
axes[0][0].set_title('High vs Low')
axes[0][0].legend()



axes[1][0].plot(high, open1, color = 'g', ls = '--',lw = 9.8 )
axes[1][0].set_xlabel('High')
axes[1][0].set_xlabel('Open')
axes[1][0].set_title('High vs Open')

axes[0][1].plot(high, close1, color = 'b', ls = '-.', lw = 5)
axes[0][1].set_xlabel('High')
axes[0][1].set_xlabel('close1')
axes[0][1].set_title('High vs close1')

axes[1][1].plot(high, volume, color = 'y', ls = '-')
axes[1][1].set_xlabel('High')
axes[1][1].set_xlabel('Volume')
axes[1][1].set_title('High vs Volume')


plt.tight_layout()





plt.show()
```

```python
apple = pd.read_csv('datasets/apple_stocks.csv')
high = np.array(apple['High'])
low = np.array(apple['Low'])
open1 = np.array(apple['Open'])
fig = plt.figure(figsize = (10,8))

ax = fig.add_axes([0, 0, 0.8, 0.8])

ax.plot(apple['High'], apple['Low'],  marker = 'o', markersize=8, markerfacecolor = 'red', markeredgecolor = 'Yellow')
ax.set_xlabel('High')
ax.set_ylabel('Low')
ax.set_title('High vs low')
ax.margins(0,0.5)

ax.legend(loc = 0, labels = ['High vs Low', 'High vs Open'])
plt.show()
```
